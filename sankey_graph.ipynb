```
import osmnx as ox
import matplotlib.pyplot as plt
```

```
gdf=ox.geocoder.geocode_to_gdf({'country':"switzerland"})
```

```
fig, ax = plt.subplots(figsize=(12, 8))
gdf.plot(ax=ax, color='blue', alpha=0.3, label='Switzerland')
```

```
G = ox.graph.graph_from_place(", CA, USA", network_type="drive")
# explore graph nodes interactively, with different basemap tiles
nodes = ox.convert.graph_to_gdfs(G, edges=False)
nodes.explore(tiles="cartodbpositron", marker_kwds={"radius": 8})
```
